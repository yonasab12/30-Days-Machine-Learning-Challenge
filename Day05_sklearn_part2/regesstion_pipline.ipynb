```python
import pandas as pd
```

```python
from sklearn .preprocessing import  OneHotEncoder,OrdinalEncoder,StandardScaler
from sklearn.model_selection import  train_test_split,cross_val_score,KFold,GridSearchCV
from  sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn .ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn .svm import SVR
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer



```

```python
x=data.drop(['Performance Index'],axis=1)
y=data['Performance Index']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=30)
```

```python
pipL = Pipeline([('oht',OneHotEncoder()),
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler(with_mean=False)),
  
    ('model', Lasso())
])
```

```python
pipL.fit(x_train, y_train)
```

```python
pipL.score(x_test, y_test)
```

```python
#pipline for linear regression
pipR = Pipeline([('oht',OneHotEncoder()),
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler(with_mean=False)),
  
    ('model', LinearRegression())
])
```

```python
pipR.fit(x_train, y_train)
```

```python
pipR.score(x_test, y_test)
```

```python
# pipline of  Riged
pipRg = Pipeline([('oht',OneHotEncoder()),
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler(with_mean=False)),
    
    ('model', Ridge())
])
```

```python
pipRg.fit(x_train, y_train)
```

```python
pipRg.score(x_test, y_test)
```

```python
pipRn = Pipeline([('oht',OneHotEncoder()),
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler(with_mean=False)),
    
    ('model', ElasticNet())
])
```

```python
pipRn.fit(x_train, y_train)
```

```python
pipRn.score(x_test, y_test)
```

```python
pipRR = Pipeline([('oht',OneHotEncoder()),
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler(with_mean=False)),
    
    ('model', RandomForestRegressor())
])
```

```python
pipRR.fit(x_train, y_train)
```

```python
pipRR.score(x_test, y_test)
```

```python
pipRD = Pipeline([('oht',OneHotEncoder()),
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler(with_mean=False)),
    
    ('model', DecisionTreeRegressor())
])
```

```python
pipRD.fit(x_train, y_train)
```

```python
pipRD.score(x_test, y_test)
```

```python
pipRV = Pipeline([('oht',OneHotEncoder()),
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler(with_mean=False)),
    #('pca', PCA(n_components=10)),
    ('model', SVR())
])
```

```python
pipRV.fit(x_train, y_train)
```

```python
pipRV.score(x_test, y_test)
```

```python
Y_pred=pipRV.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score

mse = mean_squared_error(y_test, Y_pred)
mean_absolute_error = mean_squared_error(y_test, Y_pred)

r2 = r2_score(y_test, Y_pred)

mse, mean_absolute_error, r2
```
