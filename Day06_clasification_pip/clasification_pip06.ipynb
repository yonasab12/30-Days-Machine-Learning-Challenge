```python
import pandas as pd
```

```python
from sklearn .preprocessing import  OneHotEncoder,OrdinalEncoder,StandardScaler
from sklearn.model_selection import  train_test_split,cross_val_score,KFold,GridSearchCV
from  sklearn.linear_model import LogisticRegression
from sklearn .ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn .svm import SVC
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer



```

```python
x=data.drop(['Extracurricular Activities'],axis=1)
y=data['Extracurricular Activities']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=30)
```

```python
pipL = Pipeline([('oht',OneHotEncoder()),
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler(with_mean=False)),
  
    ('model', LogisticRegression())
])
```

```python
pipL.fit(x_train, y_train)
```

```python
pipL.score(x_test, y_test)
```

```python
#pipline for linear regression
pipR = Pipeline([('oht',OneHotEncoder()),
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler(with_mean=False)),
  
    ('model', DecisionTreeClassifier())
])
```

```python
pipR.fit(x_train, y_train)
```

```python
pipR.score(x_test, y_test)
```

```python
# pipline of  Riged
pipRg = Pipeline([('oht',OneHotEncoder()),
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler(with_mean=False)),
    
    ('model', RandomForestClassifier())
])
```

```python
pipRg.fit(x_train, y_train)
```

```python
pipRg.score(x_test, y_test)
```

```python
pipRn = Pipeline([('oht',OneHotEncoder()),
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler(with_mean=False)),
    
    ('model', SVC())
])
```

```python
# hyper parameter fine tunning

from sklearn.model_selection import KFold, GridSearchCV


# Define pipeline with preprocessing and model
pipRn = Pipeline([
    ('scaler', StandardScaler()),  # Example: scaling input data
    ('model', RandomForestClassifier(random_state=30))
])

# Define parameter grid
grid_params = {
    'model__n_estimators': [100, 200, 300, 500],
    'model__max_depth': [None, 5, 10, 15],
    'model__min_samples_split': [2, 5, 10],
    'model__min_samples_leaf': [1, 2, 4]
}

# Define cross-validation scheme
kfold = KFold(n_splits=10, random_state=30, shuffle=True)

# Set up GridSearchCV
grid_search_cv = GridSearchCV(estimator=pipRn, param_grid=grid_params, cv=kfold, scoring='accuracy')

# Fit the model
grid_search_cv.fit(x_train, y_train)

# Retrieve best parameters and score
best_params = grid_search_cv.best_params_
best_score = grid_search_cv.best_score_

print("Best Parameters:", best_params)
print("Best Cross-Validation Score:", best_score)
```

```python
pipRn.fit(x_train, y_train)
```

```python
pipRn.score(x_test, y_test)
```

```python
y_pred = pipR.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix

accuracy_score = accuracy_score(y_test, y_pred)
classification_report = classification_report(y_test, y_pred)

confusion_matrix= confusion_matrix(y_test, y_pred)

accuracy_score,classification_report,confusion_matrix
```
